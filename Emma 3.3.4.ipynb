# DRILL - Descriptive Statistics and Normality

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

rand_norm = np.random.normal(20, 5, 100)
rand_chisquare = np.random.chisquare(4, 100)
rand_exp = np.random.exponential(2, 100)
rand_geo = np.random.geometric(0.2, 100)
rand_logistic = np.random.logistic(2, 1, 100)
rand_ray = np.random.rayleigh(1, 100)

plt.figure(figsize=(10, 5))
plt.subplot(231)
plt.hist(rand_norm, color='skyblue')
plt.axvline(rand_norm.mean(), color='b', linestyle='solid')
plt.axvline(rand_norm.mean()+rand_norm.std(), color='b', linestyle='dashed')
plt.axvline(rand_norm.mean()-rand_norm.std(), color='b', linestyle='dashed')
plt.title('Normal Distribution')

plt.subplot(232)
plt.hist(rand_chisquare, color='skyblue')
plt.axvline(rand_chisquare.mean(), color='b', linestyle='solid')
plt.axvline(rand_chisquare.mean()+rand_chisquare.std(), color='b', linestyle='dashed')
plt.axvline(rand_chisquare.mean()-rand_chisquare.std(), color='b', linestyle='dashed')
plt.title('Chisquare Distribution')

plt.subplot(233)
plt.hist(rand_exp, color='skyblue')
plt.axvline(rand_exp.mean(), color='b', linestyle='solid')
plt.axvline(rand_exp.mean()+rand_exp.std(), color='b', linestyle='dashed')
plt.axvline(rand_exp.mean()-rand_exp.std(), color='b', linestyle='dashed')
plt.title('Exponential Distribution')

plt.subplot(234)
plt.hist(rand_geo, color='skyblue')
plt.axvline(rand_geo.mean(), color='b', linestyle='solid')
plt.axvline(rand_geo.mean()+rand_geo.std(), color='b', linestyle='dashed')
plt.axvline(rand_geo.mean()-rand_geo.std(), color='b', linestyle='dashed')
plt.title('Geometric Distribution')

plt.subplot(235)
plt.hist(rand_geo, color='skyblue')
plt.axvline(rand_logistic.mean(), color='b', linestyle='solid')
plt.axvline(rand_logistic.mean()+rand_logistic.std(), color='b', linestyle='dashed')
plt.axvline(rand_logistic.mean()-rand_logistic.std(), color='b', linestyle='dashed')
plt.title('Logistic Distribution')

plt.subplot(236)
plt.hist(rand_ray, color='skyblue')
plt.axvline(rand_ray.mean(), color='b', linestyle='solid')
plt.axvline(rand_ray.mean()+rand_ray.std(), color='b', linestyle='dashed')
plt.axvline(rand_ray.mean()-rand_ray.std(), color='b', linestyle='dashed')
plt.title('Rayleigh Distribution')

plt.tight_layout()
plt.show()
```

Other than normal distribution, none of the other five distributions seemed to be well described by the mean and standard deviation. They are all skewed to an extend. I guess only those distributions with a somewhat symmetrical shape, similar to normal distribution, can be usefully described by mean and standard deviation.
